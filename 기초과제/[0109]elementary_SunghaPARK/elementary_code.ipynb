WLLN

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

n_trials = 2000  
true_mean = 3.5  

rolls = np.random.randint(1, 7, size=n_trials)

cumulative_means = np.cumsum(rolls) / np.arange(1, n_trials + 1)

plt.figure(figsize=(10, 6))
plt.plot(cumulative_means, label='Sample Mean (Cumulative)', color='blue')
plt.axhline(y=true_mean, color='red', linestyle='--', label='True Mean (3.5)')

plt.title('Law of Large Numbers (Dice Roll Simulation)')
plt.xlabel('Number of Trials (n)')
plt.ylabel('Mean Value')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

np.random.seed(42)

n_samples = 10000  
sample_size = 50   

lam = 1.0  
population_mean = 1 / lam
population_std = 1 / lam

samples = np.random.exponential(scale=1/lam, size=(n_samples, sample_size))
sample_means = np.mean(samples, axis=1)

plt.figure(figsize=(10, 6))

plt.hist(sample_means, bins=50, density=True, alpha=0.6, color='skyblue', label='Histogram of Sample Means')

mu = population_mean
sigma = population_std / np.sqrt(sample_size)
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 100)
plt.plot(x, stats.norm.pdf(x, mu, sigma), 'r-', lw=2, label='Theoretical Normal PDF')

plt.title(f'Central Limit Theorem Simulation (Sample Size n={sample_size})')
plt.xlabel('Sample Mean')
plt.ylabel('Density')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
import pandas as pd
from scipy import stats

df = pd.read_csv('heights.csv')

groups = df['which'].unique()
data_by_group = [df[df['which'] == group]['heights'] for group in groups]

f_statistic, p_value = stats.f_oneway(*data_by_group)

print(f"Groups: {groups}")
print(f"F-statistic: {f_statistic:.4f}")
print(f"P-value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print(": P-value < 0.05   . (   )")
else:
    print(": P-value >= 0.05     . (   )")
```
